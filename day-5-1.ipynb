```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('/kaggle/input/ipl-matches3-csv/ipl-matches (2).csv')
print(df.columns)



```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['MatchNumber'] = pd.to_numeric(df['MatchNumber'], errors='coerce')
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

df.dropna(subset=['Date', 'MatchNumber'], inplace=True)
df.plot(x='Date', y='MatchNumber', kind='line')
plt.show()
```

```python
df.plot(x='Date', y='MatchNumber', kind='scatter')
plt.show()
```

```python
df[['MatchNumber']].plot(kind='box')
plt.title('Boxplot of MatchNumber')
plt.show()
```

```python
df.plot(x='Date', y='MatchNumber', kind='bar')
plt.show()
```

```python
import seaborn as sns

sns.displot(df['MatchNumber'])
```

```python
counts = df['WinningTeam'].value_counts()

counts.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```

```python
df['WinningTeam'].head(5).value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie Plot of First 5 Rows of WinningTeam')
plt.ylabel('')
plt.show()
```
