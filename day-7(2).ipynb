### Find dates

```python
import re

dates = '''
01.04.2020
2020.04.01
hello-world
2020-04-01
2020-05-23
2020-06-11
2020-07-11
2020-08-11
2020/04/02
2020_04_04
2020_04_04
'''

pattern = re.compile(r'\b\d{4}[-/_]\d{2}[-/_]\d{2}\b')

matches = pattern.finditer(dates)
for match in matches:
    print(match.group())
```

### Find all dates in may or june months

```python
print('only dates with-or . in between in May or June')
pattern=re.compile(r'\d\d\d\d[-.]0[56][-.]\d\d')
matches = pattern.finditer(dates)
for match in matches:
    print(match.group())
```

### Title extraction using string

```python
import re

my_string = """
Mr Simpson
Mrs simpson
Mr.Brown
Ms Smith
mr.
"""

pattern = re.compile(r'(Mr|Mrs|Ms)\.?\s\w+')

matches = pattern.finditer(my_string)
for match in matches:
    print(match)
```

```python
import re

my_string = """
Mr Simpson
Mrs simpson
Mr.Brown
Ms Smith
mr.
"""

pattern = re.compile(r'\b(Mr|Mrs|Ms)\b\.?')

matches = pattern.finditer(my_string)
for match in matches:
    print(match.group())
```

### Forward fill

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['John', 'Samantha', 'Alex', 'Nina', 'Mark', 'Lilly', 'Tom', 'Olivia'],
    'Salary': [50000, 62000, np.nan, 58000, 54000, np.nan, 60000, 61000]
}

df = pd.DataFrame(data)
df_ffill = df.copy()
df_ffill['Salary'] = df_ffill['Salary'].fillna(method='ffill')

print("\nForward Fill:\n")
print(df_ffill[['Name', 'Salary']])
```

```python
import pandas as pd
import numpy as np

data = {
    'Employee': ['Aman', 'Bina', 'Charan', 'Deepa', 'Esha'],
    'Age': [25, np.nan, 30, np.nan, 28],
    'Salary': [50000, 60000, np.nan, 58000, np.nan],
    'Department': ['HR', np.nan, 'IT', 'Finance', np.nan]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
df_mean = df.copy()
df_mean['Age'] = df_mean['Age'].fillna(df_mean['Age'].mean())
df_mean['Salary'] = df_mean['Salary'].fillna(df_mean['Salary'].mean())
print("\nAfter Mean Imputation:\n", df_mean)
df_median = df.copy()
df_median['Age'] = df_median['Age'].fillna(df_median['Age'].median())
df_median['Salary'] = df_median['Salary'].fillna(df_median['Salary'].median())
print("\nAfter Median Imputation:\n", df_median)
df_mode = df.copy()
df_mode['Department'] = df_mode['Department'].fillna(df_mode['Department'].mode()[0])
print("\nAfter Mode Imputation (Department):\n", df_mode)
```

### Backward fill

```python

df_bfill = df.copy()
df_bfill['Salary'] = df_bfill['Salary'].fillna(method='bfill')
print("\nBackward Fill:\n")
print(df_bfill[['Employee', 'Salary']])
```
