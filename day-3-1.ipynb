### Sum of digits

```python
num = 13456
sum = 0

for digit_position in range(5):
    rem = num % 10
    sum = sum + rem
    num = num // 10

print("Sum of digits:", sum)
```

### Reverse the digits

```python
num = 2345
sum = 0

for digits_position in range(4):
    rem = num % 10
    sum = sum*10 + rem
    num = num // 10

print("reverse the digits:", sum)
```

### Find largest number in list

```python
numbers=[23,4,6,7,27,977]
largest=numbers[0]
for num in numbers:
    if num>largest:
        largest=num
print("largest number is:",largest)
```

### find all index at element of list appear in list

```python
students = ['P', 'A', 'A', 'A', 'P', 'P', 'A', 'P']
present_indexes = []

for index in range(len(students)):
    if students[index] == 'P':
        present_indexes.append(index)

print("Indexes where student is Present:", present_indexes)
```

### Find average value of numbers

### Find factorial using for loop

```python
num = int(input("Enter a number: "))
fact = 1

for i in range(1, num + 1):
    fact *= i

print("Factorial is:", fact)
```

### Find Prime numbers

```python
lower = int(input("Enter lower limit: "))
upper = int(input("Enter upper limit: "))

print("Prime numbers between", lower, "and", upper, "are:")

for num in range(lower, upper + 1):
    if num > 1:
        is_prime = True
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num)
```

```python
rows = int(input("Enter rows: "))

for i in range(rows, 0, -1):
    # spaces
    for s in range(rows - i):
        print(" ", end="")
    
    # ascending
    for j in range(1, i + 1):
        print(j, end="")
    
    # descending
    for k in range(i - 1, 0, -1):
        print(k, end="")
    
    print()
```

```python

def is_even(number):
    if number < 0:
        return "not allowed"
    elif number % 2 == 0:
        return "even"
    else:
        return "odd"
```
