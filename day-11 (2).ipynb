#### OneHot Encoder

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

df = pd.DataFrame({
    'color': ['red', 'blue', 'green'],
    'size': ['S', 'M', 'L']
})
features = df[['color', 'size']]
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded = encoder.fit_transform(features).astype(int)
feature_names = encoder.get_feature_names_out(['color', 'size'])

encoded_df = pd.DataFrame(encoded, columns=feature_names)
print(encoded_df)

```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
df = pd.DataFrame({
    'color': ['red', 'blue', 'green'],
    'size': ['S', 'M', 'L']
})
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded = encoder.fit_transform(df)
column_names = encoder.get_feature_names_out(['color', 'size'])
encoded_df = pd.DataFrame(encoded, columns=column_names)
print(encoded_df)
```
